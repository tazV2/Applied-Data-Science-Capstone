Is New York and Paris overrated as tourist destinations?

Introduction:

New York and Paris are the most well known and visited cities of the world by the tourists. Almost all tourists either want to visit or visited these two cities. But is there any logical reason behind it? Are these two cities really that good? Sometimes people like to follow the trend.

This study will show three key findings for:

1. Tourists

2. Business research

3. Difference between two cities


Data Collection and Usage:
New York and Paris are big cities so I took five points/locations center of the each cities.


For New York, five points are five locations such as Empire State Building, Museum of the City of New York,Yankee Stadium, Central park zoo and last but not least One World Trade Center.

Four points are in the Manhattan area but I kept one place just a little out of the Manhattan which is Yankee Stadium.


For Paris, Five points are Arc de Triomphe, Panthéon, Sacré-Cœur, Louvre Museum and Eiffel Tower. All of these points are center of the Paris.


Methodology:

The reason I am calling these places points because I search with the latitude and longitude coordinates. I have used Foursquare API with explore option. Foursquare did the explored the surrounding area of the coordinates and return the results.

The returned results were saved in Pandas(Data Analysis tool) pickle form and merged into csv file.

To check top five venues in each places, I have used frequency to see how many times same category appeared in each point.

For testing similarities and dissimilarities, Kmeans machine learning used to create clusters according to the similarities. To use the data to get meaningful results, I used one hot encoding.
